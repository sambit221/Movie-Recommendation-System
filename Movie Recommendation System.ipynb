## Get the dataset

opening u.data file i.e, 'tab' separated so sep='\t' and named the columns

```python
column_names = ["user_id","item_id", "rating","timestamp"]
df=pd.read_csv("ml-100k/u.data",sep='\t',names=column_names)
```

```python
df.shape
```

opening u.item folder i.e, | separated

```python
movies_title=pd.read_csv("ml-100k/u.item",sep='\|',header=None)
```

```python
print(movies_title.shape)
movies_title.head()
```

```python
movies_title=movies_title[[0,1]] # spliting the data
movies_title.columns=['item_id', 'title'] # naming the 2 columns
```

```python
movies_title.head()
```

```python
pd.merge(df,movies_title, on="item_id") # merging 2 data frames by item_id
```
